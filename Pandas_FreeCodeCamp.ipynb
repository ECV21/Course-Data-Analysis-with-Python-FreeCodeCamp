#Pandas - Series

A Pandas Series is a one-dimensional array-like object that can hold varios types of data, including integers, floats, python objects, etc.

1. Homogeneous data: All elements in a Series are of the same data type
2. Labels: Each element is associated with a label(index), which can be used to access the elements
3. Array-like: Support operations similar to those of Numpy array, such as slicing and vectorized operations.

#Reading external data & plotting

```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Pandas can easily read data stored in different file formats like CSV, JSON, XML or even Excel. Parsing always involves specifying the correct structure, encoding and other details. The READ_CSV method reads CSV file and accepts many parameters.

```python
#helping
pd.read_csv?
```

```python
df=pd.read_csv('btc-market-price.csv')
df.head()
```

The CSV file we're reading has only two columns: timestamp and price. It doesn't have a header, it contains whitespaces and has values separated by commas. Pandas automatically assigned the first row of data as headers, which is incorrect. We can overwrite this behaivor with the header parameter.

```python
df = pd.read_csv('btc-market-price.csv', header=None)
df.head()
```

```python
#We can then set the names of each column explicitely by setting the df.columns attribute:

df.columns = ['Timestamp', 'Price']
df.head()
```

```python
#We can see rows and columns numbers

df.shape
```

```python
#We can see the last rows

df.tail(3)
```

```python
#The type of PRICE column was correctly interpreted as float, but the TIMESTAMP
#was interpreted as a regular string(object in pandas notation):

df.dtypes
```

```python
#we can perform a vectorized operation to parse all the TIMESTAMP values as DATEtime object:

pd.to_datetime(df['Timestamp']).head()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.head()
```

```python
df.dtypes
```

```python
#The timestamp looks a lot like the index of this DataFram: date > price.
#We can change the autoincremental ID generated by pandas and use the Timestamp DS column as the index:

df.set_index('Timestamp', inplace=True)
df.head()
```

```python
df.loc['2017-09-29']
```

#Putting everthing together

And now, we've finally arrived to the final, desired version of the DataFrame parsed from our CSV file.


```python
df = pd.read_csv('btc-market-price.csv', header=None)
df.columns = ['Timestamp', 'Price']
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)
```

```python
df.head()
```

```python
df = pd.read_csv(
    'btc-market-price.csv',
    header=None,
    names=['Timestamp', 'Price'],
    index_col=0,
    parse_dates=True
)
```
